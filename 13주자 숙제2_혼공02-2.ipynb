## 전체 도서의 쪽수 구하기 - 성영 숙제 (1) 크기 정보 추출 (2) 쪽수 정보와 크기 정보를 동시에 저장

```python
def get_page_cnt(isbn): #성영 if we input 'isbn' number, the code return page numbers searching pages linked in designated url
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    # 검색 결과에서 해당 도서를 선택합니다.
    prd_info = soup.find('a', attrs={'class':'gd_name'})
    if prd_info == None:
        return ''
    # 도서 상세 페이지를 가져옵니다.
    url = 'http://www.yes24.com'+prd_info['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    # 상품 상세정보 div를 선택합니다.
    prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
    # 테이블에 있는 tr 태그를 가져옵니다.
    prd_tr_list = prd_detail.find_all('tr')
    # 쪽수가 들어 있는 th를 찾아 td에 담긴 값을 반환합니다.
    for tr in prd_tr_list:
        if tr.find('th').get_text() == '쪽수, 무게, 크기':
            return tr.find('td').get_text().split()[0]
    return ''
```

```python
def get_size(isbn): #성영 if we input 'isbn' number, the code return size searching pages linked in designated url
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    # 검색 결과에서 해당 도서를 선택합니다.
    prd_info = soup.find('a', attrs={'class':'gd_name'})
    if prd_info == None:
        return ''
    # 도서 상세 페이지를 가져옵니다.
    url = 'http://www.yes24.com'+prd_info['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    # 상품 상세정보 div를 선택합니다.
    prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
    # 테이블에 있는 tr 태그를 가져옵니다.
    prd_tr_list = prd_detail.find_all('tr')
    # 쪽수가 들어 있는 th를 찾아 td에 담긴 값을 반환합니다.
    for tr in prd_tr_list:
        if tr.find('th').get_text() == '쪽수, 무게, 크기':
            return tr.find('td').get_text().split()[4]
    return ''
```

## 숙제(2) 앞서 추출한 Book_size 정보와 기존 쪽수 정보와 함께 동시에 저장하는 Code

```python
def get_page_cnt2(isbn): #성영 if we input 'isbn' number, the code return page numbers searching pages linked in designated url
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    # 검색 결과에서 해당 도서를 선택합니다.
    prd_info = soup.find('a', attrs={'class':'gd_name'})
    if prd_info == None:
        return ''
    # 도서 상세 페이지를 가져옵니다.
    url = 'http://www.yes24.com'+prd_info['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    # 상품 상세정보 div를 선택합니다.
    prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
    # 테이블에 있는 tr 태그를 가져옵니다.
    prd_tr_list = prd_detail.find_all('tr')
    # 쪽수가 들어 있는 th를 찾아 td에 담긴 값을 반환합니다.
    for tr in prd_tr_list:
        if tr.find('th').get_text() == '쪽수, 무게, 크기':
            page_td = tr.find('td').get_text().split()[0] #성영 the element index[0] that is '쪽수' is extracted
            size_td = tr.find('td').get_text().split()[4] #성영 the element index[4] that is '크기' is extracted
            return page_td, size_td #성영 'for' syntax returns page_td & size_td in each column
    return ''
```

```python
get_page_cnt2(9791190090018)
```

```python
page_size = top10_books.apply(lambda row: get_page_cnt2(row['isbn13']), axis=1)
```

```python
page_size.name = 'page_size'
print(page_size)
```

```python
top10_with_book_combined = pd.merge(top10_books, page_size, left_index=True, right_index=True)
```

## `merge()` 함수의 매개변수

```python
df1 = pd.DataFrame({'col1': ['a','b','c'], 'col2': [1,2,3]})
df1
```

```python
df2 = pd.DataFrame({'col1': ['a','b','d'], 'col3': [10,20,30]})
df2
```

```python
pd.merge(df1, df2, on='col1')
```

```python
pd.merge(df1, df2, how='left', on='col1')
```

```python
pd.merge(df1, df2, how='right', on='col1')
```

```python
pd.merge(df1, df2, how='outer', on='col1')
```

```python
pd.merge(df1, df2, left_on='col1', right_on='col1')
```

```python
pd.merge(df1, df2, left_on='col2', right_index=True)
```
