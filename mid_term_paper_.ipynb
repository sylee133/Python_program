[문제1번] ChatGPT가 만들어준 Reference Coding

```python
while True:
    try:
        num = int(input("정수를 입력하세요 (-1 입력 시 종료): "))

        if num == -1:
            print("프로그램을 종료합니다.")
            break

        if num % 2 == 0:
            print(f"{num}은(는) 짝수입니다.")
        else:
            print(f"{num}은(는) 홀수입니다.")
    except ValueError:
        print("유효한 정수를 입력해주세요.")
```

[문제2번] 본인이 스스로 만든 Coding

```python
number = input ("숫자를 입력하세요: ")
number = int (number)

while True:
  if number == 0:
    break

  if 2<= number <= 9:
    for i in range (1,10):
        print (str(number) + "x" + str(i) + "=" + str(number * i))

  else:
    print ("2~9 사이의 수를 입력하세요")


```

[문제3번] 본인 스스로 만들어 본 Coding

```python
scoresum = 0
for i in range (1,11):
   name = str(input("학생이름을 입력하세요:"))
   score = int(input("점수를 입력하세요"))
   if score >= 60:
    print ("학생 " + name + "은 합격입니다")
   else:
    print ("학생 " + name + "은 불합격입니다")
   scoresum = int(scoresum + score)

average = int(scoresum / 10)
print("학생10명의 평균점수는 " + str(average))
```

[문제3번] ChatGPT가 만들어 준 Reference Coding

```python
students = []  # 이름과 점수를 저장할 리스트

for i in range(10):
    name = input(f"{i+1}번째 학생 이름: ")
    score = int(input(f"{name}의 점수: "))
    students.append((name, score))

print("\n결과:")
total_score = 0

for name, score in students:
    result = "합격" if score >= 60 else "불합격"
    print(f"{name}: {score}점 - {result}")
    total_score += score

average = total_score / 10
print(f"\n전체 평균 점수: {average:.2f}점")
```

[문제4번] 본인이 스스로 작성해 본 Coding

```python
passcode = input("암호를 입력하세요 (8자 이상, 숫자, 영문 대소문자 포함) / 종료하려면 'exit")
if passcode == "exit":
  print ("프로그램을 종료합니다")

elif len(passcode) >= 8 and any(char.isdigit() for char in passcode) and any(char.isupper() for char in passcode) and any(char.islower() for char in passcode):
  print ("사용가능한 비밀번호입니다")
else:
  print ("조건을 만족하지 않습니다")
```

[문제4번] ChatGPT가 만들어준 Reference coding

```python
while True:
    password = input("비밀번호를 입력하세요 (종료하려면 'exit'): ")

    if password.lower() == "exit":
        print("프로그램을 종료합니다.")
        break

    # 조건 체크
    length_ok = len(password) >= 8
    has_digit = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)

    if length_ok and has_digit and has_upper and has_lower:
        print("사용 가능한 비밀번호입니다.")
    else:
        print("조건을 만족하지 않습니다.")
```

[문제5번] 본인 스스로가 작성해 본 Coding

```python
import random
number = random.randint(1,100)

totalcount = 0

while True :
  guess = int(input("예상하는 숫자를 입력하세요"))
  countnb = 1
  totalcount += countnb
  if guess > number:
    print ("Down")
  elif guess < number:
    print ("Up")
  else:
    print ("정답입니다! 시도 횟수: " + str(totalcount) + "회")
    break
```

[문제5번] ChatGPT가 만들어 준 Reference coding

```python
import random

answer = random.randint(1, 100)
attempts = 0

while True:
    try:
        guess = int(input("숫자를 입력하세요 (1~100): "))
        attempts += 1

        if guess < 1 or guess > 100:
            print("1부터 100 사이의 숫자를 입력하세요.")
            continue

        if guess < answer:
            print("Up")
        elif guess > answer:
            print("Down")
        else:
            print(f"정답입니다! 시도 횟수: {attempts}회")
            break
    except ValueError:
        print("정수를 입력하세요.")
```
